```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
plt.title('Price US$')

list_country = ['Australia', 'USA', 'Italy', 'Mexico', 'France', 'Canada']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label=[country], marker='.')

plt.xticks(gas['Year'][::2])
plt.xlabel('Years')
plt.ylabel('US$')
plt.legend()
plt.show()
```
